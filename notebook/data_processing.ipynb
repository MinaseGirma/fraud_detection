## Encoding and scaling 

### Credit data scaled

```python
credit.describe()
```

```python
credit_feat = credit.drop('Class',axis=1)
```

```python
credit_col = credit_feat.columns
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scaler(name, data, columns):
    """
    Scale specified columns in the dataset using the specified scaler.
    
    Parameters:
    - name: Type of scaler ('standardScaler' or 'minMaxScaler').
    - data: DataFrame containing the data to be scaled.
    - columns: List of column names to scale.
    
    Returns:
    - DataFrame with scaled columns.
    """
    if name == 'standardScaler':
        scaler = StandardScaler()
    elif name == 'minMaxScaler':
        scaler = MinMaxScaler()
    else:
        raise ValueError("Unsupported scaler type. Use 'standardScaler' or 'minMaxScaler'.")
    
    data[columns] = scaler.fit_transform(data[columns])
    return data
```

```python
credit_scaled = scaler('standardScaler', credit_feat, credit_col)
```

```python
credit_scaled
```
